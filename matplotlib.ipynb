## pylot API vs OO API

```python
x = np.linspace(0,10,1000)
plt.title("sin(x) and cos(x)")
plt.plot(x,np.sin(x), color = "green", linestyle = "dashed",label="sin(x)"); #dashed nét gạch
plt.plot(x,np.cos(x), color = "red", label="cos(x)" )
plt.xlabel("X")
plt.ylabel("Sin(X)");
#plt.axis([0,4,-0.7,0.7]) #crop
#plt.axis("tight") #điều chỉnh sao cho vừa khít
plt.axis("equal") #làm hai trục bằng nhau
plt.legend(); #show plt.show()
#plt.xlim([0,5]) #Lấy khoảng 0 đến 5 
#plt.ylim([-0.7,0.6])
```

## OO API

```python
x = np.arange(1,5,1)
y = np.arange(11,45,11)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.set(title = "A simple plot",
          xlabel="x-axis",
          ylabel="y-axis");
```

### Scatter

```python
plt.scatter(x,np.exp(x)) #(x,e^x)
```

```python
#OO API

rng = np.random.RandomState(0)
x= rng.randn(100)
y= rng.randn(100)
color = rng.randn(100)
sizes = 1000*rng.rand(100)
fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s= sizes, c= color, cmap='viridis', alpha=0.3); #cmap map color, alpha độ trong suốt
fig.colorbar(img1); #mức độ màu
```

### Bar

#### vertical

```python
soft_drink_prices = {"coke":10,
                     "pessi":12,
                     "sprite":8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title = "prices drink",
      ylabel="price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(soft_drink_prices.keys(),soft_drink_prices.values()) #cột ngang
ax.set(xlabel="price ($)")
```

### Histogram

```python
np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
plt.hist(student_height, bins=30); #chia thành 30 khoảng
```

### Subplots

```python
#option1:
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows =2,
                                           ncols=2,
                                            figsize=(10,5))

ax1.plot(x,x/2);
ax1.plot(x,x*2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax4.hist(student_height, bins=30);
```

```python
#option2:
fig, ax = plt.subplots(nrows =2,ncols=2,figsize=(10,5))

ax[0,0].plot(x,x/2);
ax[0,0].plot(x,x*2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax[1,1].hist(student_height, bins=30);
```

```python
sizes = [25, 35, 20, 15, 5]  # Phần trăm các phần
labels = ['Khu vực A', 'Khu vực B', 'Khu vực C', 'Khu vực D', 'Khu vực E']  # Tên các phần

# Vẽ biểu đồ tròn
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Đặt tiêu đề cho biểu đồ
plt.title("Phân bố Dân cư theo Khu vực")

# Hiển thị biểu đồ
plt.show()
```
